# 🏢 Seção 1 — Entendimento do Negócio


## 1.1 Qual problema estamos resolvendo?

Uma empresa de e-commerce em rápido crescimento enfrenta **alta variabilidade no NPS (Net Promoter Score)** — o principal termômetro de lealdade e satisfação dos seus clientes.

O grande problema operacional é que o NPS **só é medido depois** que toda a jornada de compra já aconteceu: o pedido foi feito, entregue (ou não), o cliente ligou para o SAC, reclamou (ou não) — e só então responde à pesquisa. Nesse momento, **já é tarde para agir**. O cliente insatisfeito já viveu a experiência ruim.

A missão deste projeto é usar os **dados operacionais disponíveis em tempo real** — informações de logística, atendimento e histórico do cliente — para entender **o que causa a insatisfação** e, futuramente, **prever o NPS antes mesmo de o cliente responder**, possibilitando ações preventivas.


## 1.2 Por que o NPS é importante para um e-commerce?

O **NPS (Net Promoter Score)** é a métrica mais usada no mundo para medir lealdade de clientes. Ele parte de uma única pergunta:

> *"Em uma escala de 0 a 10, o quanto você recomendaria nossa empresa para um amigo ou familiar?"*

- **0 a 6 → Detratores:** Clientes insatisfeitos. Falam mal da empresa. Abandonam a marca.
- **7 a 8 → Neutros:** Satisfeitos, mas sem entusiasmo. Podem migrar para o concorrente.
- **9 a 10 → Promotores:** Clientes leais e entusiastas. Indicam para outros. Voltam a comprar.

**Fórmula:** `NPS = % Promotores − % Detratores`

No e-commerce, o NPS impacta diretamente:

| Dimensão | Impacto |
|---|---|
| **Recompra** | Promotores têm 5x mais probabilidade de comprar novamente |
| **Boca a boca** | Um promotor indica em média 3 novos clientes; um detrator afasta 5 |
| **Market share** | Empresas com NPS alto crescem 2x mais rápido que concorrentes |
| **Custo de aquisição** | Manter um cliente custa 5–7x menos do que conquistar um novo |
| **Receita** | Aumentar retenção em 5% pode elevar o lucro em até 95% |


## 1.3 Áreas da empresa beneficiadas pela análise

| Área | Como se beneficia |
|---|---|
| **Logística / Fulfillment** | Entende o impacto real do atraso no NPS; prioriza entregas críticas |
| **SAC / Atendimento** | Sabe quais clientes estão em risco antes que liguem; reduz volume reativo |
| **Produto / Tecnologia** | Identifica gaps na comunicação proativa (rastreamento, notificações) |
| **Precificação / Comercial** | Entende que o valor do pedido não protege a satisfação — a operação protege |
| **Estratégia / C-Level** | Dimensiona o risco de churn e impacto no crescimento orgânico |
| **Marketing** | Segmenta clientes para campanhas de reativação vs retenção |


## 1.4 Indicadores de mercado complementares

Para contextualizar o NPS da empresa, é importante conhecer benchmarks de mercado:

| Indicador | O que mede | Benchmark típico |
|---|---|---|
| **NPS — E-commerce BR** | Satisfação e lealdade geral | 40 a 60 pontos |
| **SLA Logístico** | % de entregas no prazo prometido | ≥ 95% no prazo |
| **FCR (First Contact Resolution)** | % de problemas resolvidos no 1º contato com SAC | ≥ 70% |
| **CSAT (Customer Satisfaction Score)** | Satisfação pós-atendimento | ≥ 80% |
| **CES (Customer Effort Score)** | Facilidade para resolver um problema | Quanto menor, melhor |
| **Churn Rate** | % de clientes perdidos por período | < 5% ao mês |

Esses indicadores, combinados com o NPS preditivo, formariam um **painel de saúde do cliente** que permite ação antes da perda.


## 1.5 Como o NPS impacta o negócio: Recompra, Boca a Boca e Market Share

> 💡 *Reflexão baseada no entendimento do comportamento do consumidor e nos padrões observados na análise dos dados.*

### Recompra
O NPS é um dos preditores mais fortes de recompra. Um cliente que dá nota 9 ou 10 não apenas volta — ele compra com **maior frequência** e com **ticket médio mais alto**, pois a confiança na marca já está consolidada. No cenário inverso, um detrator não só não volta: ele **cancela ativamente** recorrências e planos de fidelidade que poderia ter ativado. No dataset analisado, a `repeat_purchase_30d` está com correlação de +0.570 com o NPS, confirmando que satisfação gera retenção de forma direta e mensurável.

### Boca a Boca
O boca a boca é o canal de aquisição **mais barato e mais eficaz** que existe — e o NPS é exatamente a sua medida. Um promotor recomenda espontaneamente, sem precisar de incentivo. Um detrator exerce o oposto: relata experiências negativas para amigos, colegas e, cada vez mais, em plataformas públicas (Google, Reclame Aqui, redes sociais). Com **74% de detratores**, a empresa está gerando hoje um volume expressivo de recomendações **negativas** no mercado — elevando o custo de aquisição e prejudicando a imagem da marca em escala difícil de quantificar.

### Market Share em E-commerce
Em mercados competitivos como o e-commerce, a diferença entre ganhar e perder market share muitas vezes não está no preço — está na **experiência**. Empresas com NPS alto crescem organicamente porque seus promotores trazem novos clientes. Empresas com NPS negativo sangram: perdem clientes para concorrentes, aumentam gastos com marketing para compensar o churn e entram em um ciclo vicioso de custo crescente com base encolhendo. Com NPS de -66%, a empresa está ativamente **cedendo espaço de mercado** para concorrentes que entregam uma experiência superior — mesmo que a preços similares.

| Dimensão | Impacto com NPS Alto | Impacto com NPS Baixo (cenário atual) |
|---|---|---|
| **Recompra** | Cliente retorna com frequência e gasta mais | Cliente não retorna; abandona recorrências |
| **Boca a boca** | Recomendações espontâneas — aquisição orgânica | Críticas públicas — eleva custo de aquisição |
| **Market share** | Crescimento orgânico, menor CAC | Perda de clientes para concorrentes mais experientes |

# 🎯 Seção 2 — Definição da Variável-Alvo (Target)


## 2.1 A variável-alvo: `nps_score`

**O que é:** A nota que o cliente deu na pesquisa de satisfação NPS, em escala contínua de 0 a 10.

**Por que foi escolhida:**
- É a medida mais direta da percepção do cliente sobre toda a jornada de compra.
- Captura o resultado final de todas as interações: entrega, atendimento, produto.
- É amplamente usada pelo mercado, permitindo comparação com benchmarks externos.
- É acionável: uma queda no NPS indica onde e quando intervir.

**Momento de coleta:** Após a conclusão da jornada de compra (entrega realizada). Isso significa que, no momento da coleta, **todos os eventos operacionais já ocorreram** — o atraso, os contatos com SAC, as reclamações. A variável-alvo reflete o acúmulo de todas essas experiências.

## 2.2 Riscos de uso inadequado

| Risco | Descrição | Mitigação |
|---|---|---|
| **Viés de resposta** | Clientes muito satisfeitos ou muito insatisfeitos respondem mais. A amostra pode não representar todos os clientes. | Combinar NPS com CSAT operacional. |
| **Data leakage — `csat_internal_score`** | Essa nota interna já "sabe" o NPS que será dado — incluí-la no modelo seria trapacear. | Excluir da modelagem preditiva. |
| **Data leakage — `repeat_purchase_30d`** | A recompra em 30 dias ocorre após a nota e é consequência da satisfação, não causa. | Excluir da modelagem preditiva. |
| **Correlação ≠ causalidade** | Saber que atraso correlaciona com NPS baixo não prova que reduzir atraso vai aumentar o NPS — mas é a hipótese mais forte e testável. | Implementar experimentos controlados. |
| **Uso para punição** | Usar o NPS preditivo para punir operadores ou entregadores cria distorção e desonestidade nos dados. | Usar apenas para melhoria de processo. |

> Vamos "conversar" com os dados: entender como o NPS está distribuído, quais variáveis têm mais relação com a satisfação do cliente, onde estão os pontos de ruptura e como os problemas se encadeiam. Cada célula de código virá com explicações em linguagem simples.


```python
# ============================================================
# 📦 IMPORTANDO AS FERRAMENTAS QUE V�amos USAR
# ============================================================
# Pense nisso como "abrir a caixa de ferramentas" antes de começar o trabalho.
# Cada biblioteca tem uma função específica:

import pandas as pd        # pandas: lê e manipula tabelas de dados (como Excel, só que em Python)
import numpy as np         # numpy: faz cálculos matemáticos e estatísticos
import matplotlib.pyplot as plt  # matplotlib: desenha os gráficos
import seaborn as sns      # seaborn: deixa os gráficos mais bonitos e profissionais
import warnings            # warnings: evita que mensagens técnicas desnecessárias apareçam
warnings.filterwarnings('ignore')  # silencia avisos irrelevantes

# Configurações visuais globais dos gráficos
plt.rcParams['figure.dpi'] = 120          # resolução dos gráficos (maior = mais nítido)
plt.rcParams['font.family'] = 'sans-serif'  # fonte limpa e profissional
sns.set_style("whitegrid")                # fundo branco com linhas de grade sutis

print("✅ Bibliotecas importadas com sucesso!")
print("   pandas, numpy, matplotlib e seaborn estão prontos.")
```

```python
# ============================================================
# 📂 CARREGANDO OS DADOS
# ============================================================
# Aqui lemos o arquivo CSV com os dados dos clientes.
# sep=';' indica que as colunas são separadas por ponto e vírgula (padrão BR).
# decimal=',' indica que os decimais usam vírgula (ex: 1.234,56 → 1234.56).

df = pd.read_csv('desafio_nps_fase_1.csv')

# Exibindo as primeiras 5 linhas para ter uma "fotografia" dos dados
print("=" * 60)
print("📊 PRIMEIRAS LINHAS DO DATASET")
print("=" * 60)
display(df.head())

print(f"\n✅ Dataset carregado com sucesso!")
print(f"   📏 Dimensões: {df.shape[0]:,} linhas × {df.shape[1]} colunas")
```

```python
# ============================================================
# 🔍 EXPLORAÇÃO INICIAL — RESUMO ESTATÍSTICO
# ============================================================
# O método .describe() nos dá um resumo estatístico de cada coluna numérica:
# - count: quantos valores existem (útil para detectar nulos)
# - mean: média
# - std: desvio padrão (quanto os valores variam em torno da média)
# - min/max: valores mínimo e máximo
# - 25%/50%/75%: percentis (onde estão 25%, 50% e 75% dos dados)

print("=" * 60)
print("📊 RESUMO ESTATÍSTICO DO DATASET")
print("=" * 60)
display(df.describe().round(2))
```

```python
# ============================================================
# 🧹 VERIFICAÇÃO DE QUALIDADE DOS DADOS
# ============================================================
# Antes de qualquer análise, precisamos saber se os dados estão "limpos":
# - Valores nulos (faltando) podem distorcer cálculos.
# - Registros duplicados podem inflar contagens e médias.

nulos = df.isnull().sum().sum()           # soma todos os valores ausentes
duplicados = df.duplicated().sum()        # conta linhas idênticas

print("=" * 60)
print("🧹 VERIFICAÇÃO DE QUALIDADE DOS DADOS")
print("=" * 60)

if nulos == 0:
    print(f"  ✅ Valores nulos: {nulos} — Dataset sem dados faltantes!")
else:
    print(f"  ❌ Valores nulos encontrados: {nulos}")

if duplicados == 0:
    print(f"  ✅ Registros duplicados: {duplicados} — Nenhuma linha repetida!")
else:
    print(f"  ❌ Registros duplicados encontrados: {duplicados}")

print()
print("  📋 Colunas e tipos de dados:")
print("-" * 40)
for col, dtype in zip(df.columns, df.dtypes):
    print(f"  • {col:<35} → {str(dtype)}")
```

```python
# ============================================================
# 🏷️ CLASSIFICANDO OS CLIENTES EM DETRATOR / NEUTRO / PROMOTOR
# ============================================================
# A metodologia NPS divide os clientes em 3 grupos conforme a nota dada:
#   0 a 6  → Detratores (insatisfeitos, falam mal da empresa)
#   7 a 8  → Neutros   (satisfeitos, mas sem entusiasmo)
#   9 a 10 → Promotores (leais, indicam para outros)

def classificar_nps(nota):
    """Função que recebe a nota e retorna a classificação NPS."""
    if nota <= 6:
        return 'Detrator'
    elif nota <= 8:
        return 'Neutro'
    else:
        return 'Promotor'

# Criamos uma nova coluna aplicando a função a cada linha
df['nps_class'] = df['nps_score'].apply(classificar_nps)

# Contando quantos clientes em cada grupo
contagem = df['nps_class'].value_counts()
percentuais = df['nps_class'].value_counts(normalize=True) * 100

# Calculando o NPS da empresa
pct_promotores = percentuais.get('Promotor', 0)
pct_detratores = percentuais.get('Detrator', 0)
nps_empresa = pct_promotores - pct_detratores

print("=" * 60)
print("🏷️  CLASSIFICAÇÃO NPS DOS CLIENTES")
print("=" * 60)
print(f"  🔴 Detratores  (nota 0-6): {contagem.get('Detrator',0):>5} clientes  ({percentuais.get('Detrator',0):.1f}%)")
print(f"  🟡 Neutros     (nota 7-8): {contagem.get('Neutro',0):>5} clientes  ({percentuais.get('Neutro',0):.1f}%)")
print(f"  🟢 Promotores  (nota 9-10): {contagem.get('Promotor',0):>4} clientes  ({percentuais.get('Promotor',0):.1f}%)")
print("-" * 60)
print(f"  📊 NPS DA EMPRESA: {nps_empresa:.0f}%")
print()
if nps_empresa < 0:
    print("  🚨 ZONA CRÍTICA! NPS negativo indica que há MUITO mais")
    print("     clientes insatisfeitos do que satisfeitos.")
    print()
    print("  📌 Benchmark e-commerce Brasil: NPS entre 40% e 60%")
    print(f"  📌 Nossa empresa está {abs(nps_empresa - 40):.0f} pontos ABAIXO do mínimo aceitável!")
```

```python
# ============================================================
# 📊 VISUALIZANDO A DISTRIBUIÇÃO DO NPS
# ============================================================
# Dois gráficos complementares:
# 1. Histograma: mostra como as notas 0-10 estão distribuídas (frequência)
# 2. Pizza: mostra a proporção entre Detratores, Neutros e Promotores

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ----- GRÁFICO 1: Histograma das notas -----
cores_hist = []
for nota in sorted(df['nps_score'].round().unique()):
    if nota <= 6:
        cores_hist.append('#e74c3c')   # vermelho = detrator
    elif nota <= 8:
        cores_hist.append('#f39c12')   # amarelo = neutro
    else:
        cores_hist.append('#27ae60')   # verde = promotor

ax1 = axes[0]
counts, bins, patches = ax1.hist(df['nps_score'], bins=20, edgecolor='white', linewidth=0.8, color='#3498db')

# Recolorindo as barras por faixa
for patch in patches:
    centro = (patch.get_x() + patch.get_width() / 2)
    if centro <= 6.5:
        patch.set_facecolor('#e74c3c')
    elif centro <= 8.5:
        patch.set_facecolor('#f39c12')
    else:
        patch.set_facecolor('#27ae60')

ax1.set_title('Distribuição das Notas NPS\n(0 = péssimo, 10 = excelente)', fontsize=14, fontweight='bold', pad=15)
ax1.set_xlabel('Nota NPS', fontsize=12)
ax1.set_ylabel('Número de clientes', fontsize=12)
ax1.axvline(x=6.5, color='gray', linestyle='--', alpha=0.7, label='Limite Detrator/Neutro')
ax1.axvline(x=8.5, color='gray', linestyle=':', alpha=0.7, label='Limite Neutro/Promotor')
ax1.legend(fontsize=10)

# Anotações das zonas
ax1.text(3.0, ax1.get_ylim()[1]*0.85, 'DETRATORES\n(0-6)', ha='center', fontsize=10,
         color='#c0392b', fontweight='bold')
ax1.text(7.5, ax1.get_ylim()[1]*0.85, 'NEUTROS\n(7-8)', ha='center', fontsize=10,
         color='#e67e22', fontweight='bold')
ax1.text(9.5, ax1.get_ylim()[1]*0.85, 'PROM.\n(9-10)', ha='center', fontsize=10,
         color='#1e8449', fontweight='bold')

# ----- GRÁFICO 2: Pizza das categorias -----
ax2 = axes[1]
categorias = ['Detratores\n(0-6)', 'Neutros\n(7-8)', 'Promotores\n(9-10)']
valores = [contagem.get('Detrator',0), contagem.get('Neutro',0), contagem.get('Promotor',0)]
cores_pizza = ['#e74c3c', '#f39c12', '#27ae60']
explode = (0.05, 0, 0.05)  # destaca levemente Detratores e Promotores

wedges, texts, autotexts = ax2.pie(
    valores, labels=categorias, autopct='%1.1f%%',
    colors=cores_pizza, explode=explode,
    textprops={'fontsize': 11}, startangle=90
)
for at in autotexts:
    at.set_fontweight('bold')
    at.set_fontsize(12)

ax2.set_title(f'Composição do NPS\nNPS = {nps_empresa:.0f}% (Zona Crítica 🚨)', 
              fontsize=14, fontweight='bold', pad=15)

plt.suptitle('Distribuição do NPS — E-commerce | n = 2.500 clientes', 
             fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('fig01_distribuicao_nps.png', bbox_inches='tight', dpi=150)
plt.show()

print()
print("=" * 60)
print("💡 INSIGHT — DIAGNÓSTICO GERAL:")
print("=" * 60)
print("  🚨 Com NPS de -66%, a empresa está em ZONA CRÍTICA.")
print("  🔴 74% dos clientes são Detratores — falam MAL da empresa.")
print("  🟢 Apenas 8% são Promotores — precisamos multiplicar esse número.")
print()
print("  📌 Benchmark e-commerce BR: NPS entre 40% e 60%.")
print("  📌 Empresas com NPS negativo tendem a PERDER market share")
print("     rapidamente para concorrentes que entregam experiência.")
print("=" * 60)
```

```python
# ============================================================
# 🔗 CORRELAÇÕES COM O NPS — O QUE MAIS IMPACTA A NOTA?
# ============================================================
# A correlação de Pearson mede a força da relação LINEAR entre duas variáveis.
# Varia de -1 a +1:
#   +1 = quando X sobe, Y sobe junto (correlação perfeita positiva)
#    0 = sem relação linear detectada
#   -1 = quando X sobe, Y cai (correlação perfeita negativa)
#
# Regra de bolso para interpretar:
#   |r| > 0.5 → forte  |  0.3-0.5 → moderada  |  < 0.3 → fraca

# Calculando correlações de todas as variáveis numéricas com o nps_score
correlacoes = df.corr(numeric_only=True)['nps_score'].drop('nps_score')

# Removendo IDs (não fazem sentido estatístico)
correlacoes = correlacoes.drop(['customer_id', 'order_id'], errors='ignore')

# Ordenando do mais negativo para o mais positivo
correlacoes_ordenadas = correlacoes.sort_values()

# Tabela formatada
print("=" * 65)
print("📊 RANKING DE CORRELAÇÃO COM NPS (Pearson)")
print("=" * 65)
print(f"  {'Variável':<35} {'Correlação':>12}  {'Força'}")
print("-" * 65)

for var, r in correlacoes_ordenadas.items():
    if abs(r) >= 0.5:
        forca = "🔴 FORTE" if r < 0 else "🟢 FORTE"
    elif abs(r) >= 0.3:
        forca = "🟡 Moderada"
    else:
        forca = "⬜ Fraca / irrelevante"
    sinal = "↓" if r < 0 else "↑"
    print(f"  {var:<35} {r:>+.3f}  {sinal}   {forca}")

print("=" * 65)
```

```python
# ============================================================
# 📊 GRÁFICO DE CORRELAÇÕES COM O NPS
# ============================================================

fig, ax = plt.subplots(figsize=(14, 7))

# Definindo cores: vermelho para correlações negativas, verde para positivas
cores = ['#e74c3c' if r < 0 else '#27ae60' for r in correlacoes_ordenadas.values]

# Criando barras horizontais
bars = ax.barh(
    y=correlacoes_ordenadas.index,
    width=correlacoes_ordenadas.values,
    color=cores,
    edgecolor='white',
    height=0.7
)

# Anotando os valores nas barras
for bar, valor in zip(bars, correlacoes_ordenadas.values):
    offset = 0.01 if valor >= 0 else -0.01
    ha = 'left' if valor >= 0 else 'right'
    ax.text(valor + offset, bar.get_y() + bar.get_height() / 2,
            f'{valor:+.3f}', va='center', ha=ha, fontsize=9, fontweight='bold')

# Linha vertical no zero
ax.axvline(x=0, color='black', linewidth=1.2)

# Linhas de referência para faixas de força
ax.axvline(x=0.5, color='green', linewidth=0.8, linestyle='--', alpha=0.5, label='Forte positivo (≥0.5)')
ax.axvline(x=-0.5, color='red', linewidth=0.8, linestyle='--', alpha=0.5, label='Forte negativo (≤-0.5)')

ax.set_title('Correlação de Pearson com NPS Score\n(variáveis ordenadas da mais negativa para a mais positiva)',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Coeficiente de Correlação de Pearson (r)', fontsize=12)
ax.set_xlim(-0.75, 0.75)
ax.legend(fontsize=10, loc='lower right')

# Rótulo lateral
ax.text(0.62, 0.95, 'Verde = melhora NPS\nVermelho = piora NPS', transform=ax.transAxes,
        fontsize=9, verticalalignment='top', 
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

plt.tight_layout()
plt.savefig('fig02_correlacoes_nps.png', bbox_inches='tight', dpi=150)
plt.show()

print()
print("=" * 65)
print("💡 INSIGHT — OS 5 FATORES QUE MAIS IMPACTAM O NPS:")
print("=" * 65)
print()
print("  🔴 PIORES para o NPS (correlação NEGATIVA):")
print("     1. delivery_delay_days  (-0.597): Atraso na entrega é o PRINCIPAL")
print("        vilão — quando sobe, o NPS despenca fortemente.")
print("     2. complaints_count     (-0.497): Cada reclamação corrói a nota.")
print("     3. customer_service_contacts (-0.351): Mais contatos com SAC = cliente")
print("        frustrado tentando resolver problemas.")
print()
print("  🟢 MELHORES para o NPS (correlação POSITIVA):")
print("     4. repeat_purchase_30d  (+0.570): Clientes satisfeitos voltam a comprar.")
print("        ⚠️  ATENÇÃO: não usar como preditor (data leakage).")
print("     5. csat_internal_score  (+0.564): Nota interna já reflete satisfação.")
print("        ⚠️  ATENÇÃO: não usar como preditor (data leakage).")
print()
print("  ⬜ IRRELEVANTES para o NPS: idade, região, tempo como cliente,")
print("     valor do pedido, desconto. O NPS não discrimina por perfil")
print("     demográfico — discrimina por QUALIDADE OPERACIONAL.")
print("=" * 65)
```

```python
# ============================================================
# ⚡ PONTO DE RUPTURA: NPS POR FAIXA DE ATRASO
# ============================================================
# Este é o GRÁFICO MAIS IMPORTANTE de toda a análise.
# Vamos ver exatamente em qual dia de atraso o NPS "despenca"
# e qual a proporção de clientes detratores em cada faixa.

# Classificando o atraso em faixas de dias
def faixa_atraso(dias):
    if dias == 0:
        return '0 dias\n(sem atraso)'
    elif dias <= 2:
        return '1-2 dias'
    elif dias <= 4:
        return '3-4 dias\n⚡ RUPTURA'
    else:
        return '5-8 dias\n🚨 CRÍTICO'

df['faixa_atraso'] = df['delivery_delay_days'].apply(faixa_atraso)

# Calculando NPS médio e % de detratores por faixa
analise_atraso = df.groupby('faixa_atraso').agg(
    nps_medio=('nps_score', 'mean'),
    pct_detratores=('nps_class', lambda x: (x == 'Detrator').mean() * 100),
    total=('nps_score', 'count')
).reset_index()

# Ordenando pela lógica de gravidade
ordem = ['0 dias\n(sem atraso)', '1-2 dias', '3-4 dias\n⚡ RUPTURA', '5-8 dias\n🚨 CRÍTICO']
analise_atraso['faixa_atraso'] = pd.Categorical(analise_atraso['faixa_atraso'], categories=ordem, ordered=True)
analise_atraso = analise_atraso.sort_values('faixa_atraso')

# Tabela resumo
print("=" * 65)
print("⚡ NPS E % DE DETRATORES POR FAIXA DE ATRASO")
print("=" * 65)
print(f"  {'Faixa':<25} {'NPS Médio':>10} {'% Detratores':>14} {'Clientes':>10}")
print("-" * 65)
for _, row in analise_atraso.iterrows():
    faixa = row['faixa_atraso'].replace('\n', ' ')
    print(f"  {faixa:<25} {row['nps_medio']:>10.2f} {row['pct_detratores']:>13.1f}% {int(row['total']):>10}")
print("=" * 65)
```

```python
# ============================================================
# 📊 NPS POR RECLAMAÇÕES E CONTATOS COM SAC
# ============================================================
# Agora analisamos como o acúmulo de reclamações e contatos com
# o atendimento ao cliente impactam o NPS.

def faixa_reclamacoes(n):
    if n == 0:   return '0 reclamações'
    elif n <= 2: return '1-2 reclamações'
    elif n <= 4: return '3-4 reclamações'
    else:        return '5+ reclamações'

df['faixa_reclam'] = df['complaints_count'].apply(faixa_reclamacoes)

# NPS por faixa de reclamações
nps_reclam = df.groupby('faixa_reclam')['nps_score'].mean()
ordem_r = ['0 reclamações', '1-2 reclamações', '3-4 reclamações', '5+ reclamações']
nps_reclam = nps_reclam.reindex(ordem_r)

# NPS por contatos SAC (agrupando 4+)
df['sac_group'] = df['customer_service_contacts'].apply(
    lambda x: '4+ contatos' if x >= 4 else f'{x} contato(s)'
)
ordem_s = ['0 contato(s)', '1 contato(s)', '2 contato(s)', '3 contato(s)', '4+ contatos']
nps_sac = df.groupby('sac_group')['nps_score'].mean().reindex(ordem_s)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# ----- GRÁFICO 1: NPS por reclamações -----
ax1 = axes[0]
cores_r = ['#27ae60', '#f39c12', '#e74c3c', '#922b21']
barras_r = ax1.bar(range(len(nps_reclam)), nps_reclam.values, color=cores_r,
                    edgecolor='white', width=0.6)
for b, v in zip(barras_r, nps_reclam.values):
    ax1.text(b.get_x() + b.get_width()/2, v + 0.1, f'{v:.2f}',
             ha='center', fontsize=12, fontweight='bold')
ax1.set_xticks(range(len(nps_reclam)))
ax1.set_xticklabels(nps_reclam.index, fontsize=10, rotation=10)
ax1.set_ylabel('NPS Médio', fontsize=12)
ax1.set_title('NPS Médio por\nQuantidade de Reclamações', fontsize=13, fontweight='bold')
ax1.set_ylim(0, 11)
ax1.yaxis.grid(True, alpha=0.4)

# ----- GRÁFICO 2: NPS por contatos SAC -----
ax2 = axes[1]
cores_s = ['#27ae60', '#a9cce3', '#f39c12', '#e74c3c', '#922b21']
barras_s = ax2.bar(range(len(nps_sac)), nps_sac.values, color=cores_s,
                    edgecolor='white', width=0.6)
for b, v in zip(barras_s, nps_sac.values):
    ax2.text(b.get_x() + b.get_width()/2, v + 0.1, f'{v:.2f}',
             ha='center', fontsize=12, fontweight='bold')
ax2.set_xticks(range(len(nps_sac)))
ax2.set_xticklabels(nps_sac.index, fontsize=10, rotation=10)
ax2.set_ylabel('NPS Médio', fontsize=12)
ax2.set_title('NPS Médio por Número de\nContatos com SAC', fontsize=13, fontweight='bold')
ax2.set_ylim(0, 11)
ax2.yaxis.grid(True, alpha=0.4)

plt.suptitle('Impacto das Reclamações e do SAC no NPS', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('fig04_reclam_sac.png', bbox_inches='tight', dpi=150)
plt.show()

print()
print("=" * 65)
print("💡 INSIGHTS — RECLAMAÇÕES E SAC:")
print("=" * 65)
print()
print("  📌 RECLAMAÇÕES:")
print("     • 0 reclamações → NPS 8.52  (clientes satisfeitos!)")
print("     • 1-2 reclamações → NPS 6.58  (ainda recuperável)")
print("     • 3-4 reclamações → NPS 4.54  (sinal de alarme)")
print("     • 5+ reclamações → NPS 3.20  (cliente praticamente perdido)")
print()
print("  📌 CONTATOS SAC:")
print("     • 0 contatos → NPS 5.54")
print("     • 1 contato  → NPS 4.66  (-16% de queda)")
print("     • 2 contatos → NPS 4.12  (-12% adicional)")
print("     • 3 contatos → NPS 3.20  (-22% adicional)")
print("     • 4+ contatos → NPS < 2.7  (queda acentuada)")
print()
print("  ➡️  Cada contato adicional com o SAC é um sinal de FALHA.")
print("     O cliente não deveria precisar ligar várias vezes para resolver")
print("     o mesmo problema. Resolver na PRIMEIRA ligação é fundamental.")
print("=" * 65)
```

```python
# ============================================================
# 🌊 EFEITO CASCATA: ATRASO → RECLAMAÇÕES → SAC → NPS
# ============================================================
# Um dos descobertas mais importantes da análise:
# O atraso na entrega NÃO impacta só o NPS diretamente.
# Ele desencadeia uma CASCATA de problemas que, juntos, destroem a satisfação.
#
# Modelo da cascata:
# [ATRASO] → gera mais [RECLAMAÇÕES] → gera mais [CONTATOS SAC] → derruba o [NPS]

corr_atraso_reclam = df['delivery_delay_days'].corr(df['complaints_count'])
corr_reclam_sac    = df['complaints_count'].corr(df['customer_service_contacts'])
corr_sac_nps       = df['customer_service_contacts'].corr(df['nps_score'])
corr_atraso_nps    = df['delivery_delay_days'].corr(df['nps_score'])

print("=" * 65)
print("🌊 ANÁLISE DO EFEITO CASCATA")
print("=" * 65)
print()
print("  [ATRASO] ──────────────────────────────► [NPS]")
print(f"               correlação direta: {corr_atraso_nps:.3f} 🔴")
print()
print("  [ATRASO] ──► [RECLAMAÇÕES] ──► [SAC] ──► [NPS]")
print(f"                r={corr_atraso_reclam:.3f}         r={corr_reclam_sac:.3f}    r={corr_sac_nps:.3f}")
print()
print("  Interpretação da cadeia:")
print(f"  1. Atraso gera reclamações:   r = {corr_atraso_reclam:.3f} (positivo — sobem juntos)")
print(f"  2. Reclamações geram SAC:     r = {corr_reclam_sac:.3f} (muito forte!)")
print(f"  3. SAC impacta NPS:           r = {corr_sac_nps:.3f} (negativo — SAC sobe, NPS cai)")
print()
print("-" * 65)
print("  ⚡ CONCLUSÃO ESTRATÉGICA:")
print()
print("  Quando a logística atrasa, o cliente reclama. Quando reclama,")
print("  liga para o SAC. Quando precisa ligar várias vezes, o NPS despenca.")
print()
print("  RESOLVER O ATRASO = QUEBRAR A CASCATA INTEIRA.")
print("  Uma única ação (melhorar a entrega) reduz reclamações, reduz")
print("  chamadas no SAC e eleva o NPS — três ganhos pelo preço de um.")
print("=" * 65)
```

# 🤖 Seção 4 — Reflexão sobre Modelagem Preditiva

> **Atenção:** Esta seção é **conceitual**. Não implementamos nenhum modelo aqui. O objetivo é refletir sobre como a Inteligência Artificial poderia usar os dados que analisamos para prever a satisfação do cliente **antes** de ele responder à pesquisa.


## 4.1 Duas abordagens possíveis

### Abordagem A — Regressão (prever um número)
O modelo tentaria prever **exatamente qual nota** (0 a 10) o cliente dará.  
- **Vantagem:** mais granular, permite simulações finas.  
- **Desvantagem:** prever "4.2 vs 4.8" tem pouco valor prático para a operação.

### Abordagem B — Classificação (prever um grupo)
O modelo tentaria prever **em qual categoria** o cliente vai cair: Detrator, Neutro ou Promotor.  
- **Vantagem:** diretamente acionável — "este cliente tem 87% de chance de virar detrator".  
- **Desvantagem:** perde a granularidade da nota exata.

---

## 4.2 Recomendação: Classificação Binária (Detrator vs Não-Detrator)

Para uso operacional imediato, a abordagem mais eficaz seria uma **classificação binária**: simplesmente prever se o cliente **vai ou não ser detrator**.

Por quê?  
- É a pergunta que a operação consegue agir: "este cliente está em risco?"  
- Gera um **score de risco** por pedido, em tempo real.  
- Pode ser integrada a alertas automáticos nos sistemas logísticos e de SAC.


## 4.3 Variáveis de entrada recomendadas

Para a modelagem preditiva, usaríamos apenas variáveis disponíveis **no momento do pedido** ou durante a operação — nunca variáveis que "já sabem" o resultado:

| Variável | Inclui? | Motivo |

## 4.6 Justificativa Técnica e de Negócio da Estratégia Escolhida

### Por que Classificação Binária é a escolha certa para este negócio?

**Do ponto de vista técnico:**
- O `nps_score` contínuo (0 a 10) apresenta alta variabilidade dentro de cada grupo operacional, o que dificulta a precisão de modelos de regressão.
- A classificação binária (Detrator vs Não-Detrator) reduz o ruído e foca o modelo no que importa: **identificar quem está em risco antes que o problema aconteça**.
- Com 74% de detratores na base, o modelo tem volume suficiente para aprender os padrões de insatisfação com confiança estatística.
- As variáveis preditoras são numéricas e contínuas — adequadas para a maioria dos algoritmos de classificação supervisionada disponíveis.

**Do ponto de vista de negócio:**
- A operação não precisa saber se o cliente vai dar nota 3.2 ou 3.8 — ela precisa saber se ele **vai ou não reclamar e falar mal da empresa**.
- A classificação binária permite criar um **alerta acionável e direto**: "Pedido X está em risco alto — ativar protocolo de intervenção."
- É mais fácil de comunicar para equipes não-técnicas: *"Este cliente tem 85% de chance de virar detrator"* é uma frase que o gerente de logística entende e consegue agir imediatamente.
- Reduz o custo de treinamento, manutenção e monitoramento do modelo no longo prazo.
